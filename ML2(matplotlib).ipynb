```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.arange(1,10)
y=np.arange(11,20)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title('scatter plot')
plt.savefig('scat.png')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=20)
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title('plot')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,2)
plt.plot(x,y,'b*',linestyle='dashed',linewidth=2,markersize=20)
plt.subplot(2,2,4)
plt.scatter(x,y,c='r')
plt.subplot(2,2,3)
plt.plot(x,y,'b*',linestyle='dashed',linewidth=1,markersize=10)
```

```python
labels='python','c++','Java','rubby'
size=[215,150,230,675]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
```

```python
plt.pie(size,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
```

```python
explode=(0.3,0,0,0)
plt.pie(size,labels=labels,explode=explode,colors=colors,autopct='%1.1f%%',shadow=True)
```
